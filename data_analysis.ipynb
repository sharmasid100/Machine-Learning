```python
df = df[['coaster_name', 
        #'Length', 'Speed',
        'Location', 'Status',
        #'Opening date',
       #'Type', 
        
    'Manufacturer',

       #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
        
       'year_introduced',
        'latitude', 'longitude',
        'Type_Main',
       'opening_date_clean', 
        #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
        #'height_value', 'height_unit', 
        'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df
```

```python
df.dtypes
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.isna().sum()
```

```python
#REMOVE DUPLICATES
df = df.loc[~df.duplicated(subset = ['coaster_name','Location','opening_date_clean'])].reset_index(drop = True)
```

```python
#FEATURE ANALYSIS
```

```python
ax1 = df['year_introduced'].value_counts().head(10).plot(kind = 'bar' , title = "Year of introduction")
ax1.set_xlabel('Year')
ax1.set_ylabel('count')
```

```python
ax2 = df['speed_mph'].plot(kind = 'hist', title = 'speed')
ax2.set_xlabel('mph')
ax2.set_ylabel('count')
```

```python
#FEATURE RELATIONSHIP
```

```python
ax3 = df.plot(kind = 'scatter' , x = 'speed_mph' , y = 'height_ft' , title = 'speed vs height')
ax3.set_xlabel('speed')
ax3.set_ylabel('height')
plt.show()
```

```python
sns.scatterplot(data = df , x = 'speed_mph' , y = 'height_ft' , hue = 'year_introduced')
plt.show()
```

```python
df.columns
```

```python
sns.pairplot(data = df , kind = 'scatter' , vars =['year_introduced',
       'latitude', 'longitude', 'opening_date_clean', 'speed_mph',
       'height_ft'] , markers = '*')
plt.show()
```
